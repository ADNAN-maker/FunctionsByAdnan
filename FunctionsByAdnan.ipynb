Q3. What are the different ways to define and call a function in Python?

---
Ans. 1. Standard Function Definition
using the def keyword, followed by the function name, parentheses for arguments, and a colon.
2. Lambda Functions are anonymous, small, and defined inline using the lambda keyword. They are limited to a single expression.
3. Nested Functions: function can be defined inside other functions. The inner function has access to the outer function's variables (closure).
4. Recursive Functions: A function can call itself, which is known as recursion. This is useful for problems that can be broken down into smaller, self-similar subproblems.
5. Functions with Arbitrary Arguments (*args and **kwargs)
*   *args allows a function to accept any number of positional arguments, which are collected into a tuple.
*   **kwargs allows a function to accept any number of keyword arguments, which are collected into a dictionary.
6. Callable Objects with __call__
Classes can be made callable by defining the __call__ method. This allows instances of the class to be called like functions.

```python
# example for functions in which differnt ways of define and call was performed.
def greet(name):
    print("standard function exemple:",(f"Hello, {name}!"))
greet("Adnan")
#------------------------------------------------------------------
add = lambda x, y: x + y
print("lambda function example:",(add(5, 3)))
#------------------------------------------------------------------
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function
add_five = outer_function(5)
print("recursive Function example:",(add_five(3)))
#------------------------------------------------------------------
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
print("Functions with Arbitrary Arguments:",(factorial(5)))
#------------------------------------------------------------------
def print_arguments(*args, **kwargs):
    for arg in args:
        print(f"Argument: {arg}")
    for key, value in kwargs.items():
        print(f"Keyword argument: {key} = {value}")
print_arguments(1, 2, 3, name="Adnan", age=29)
#------------------------------------------------------------------
class Greeter:
    def __call__(self, name):
        print(f"Greetings, {name}!")
greeter = Greeter()
greeter("Adnan")
```

Q6. Explain the concept of generators in Python and how they are defined.

---
Ans. A generator function is a special type of function that returns an iterator object. Instead of using return to send back a single value, generator functions use yield to produce a series of results over time. This allows the function to generate values and pause its execution after each yield, maintaining its state between iterations.


```python
# example of Generator
def pain(max):
    relif = 1
    while relif <= max:
        yield relif
        relif += 1

relif = pain(5)
for n in relif:
    print(n)
```

Q7. What are the advantages of using generators over regular functions?

---
Ans. Advantages over regular functions
1. Lazy Evaluation:
Generators and iterators use lazy evaluation, meaning they generate values on-the-fly and don’t store all values in memory at once. This is particularly useful when dealing with large datasets or infinite sequences.
2. Memory Efficiency:
Traditional for loops can consume a lot of memory if they generate a large list. Generators, on the other hand, generate values one at a time, conserving memory.
3. Code Simplicity:
Generators provide a more concise and readable way to create iterators. They eliminate the need to manage the state of the iteration manually, making the code cleaner.
4. Infinite Sequences:
Generators allow you to represent infinite sequences efficiently, as they generate values as needed without the need to precompute and store the entire sequence.

Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

---
Ans. **The map()** function in Python is a built-in function used to apply a function to every item in an iterable (like a list, tuple, etc.) and return a new iterable (a map object) with the results.

**reduce()** works differently than map() and filter(). It does not return a new list based on the function and iterable we've passed. Instead, it returns a single value. It performs a rolling computation on pairs of elements.

**filter()** takes a function object and an iterable and creates a new list.
As the name suggests, filter() forms a new list that contains only elements that satisfy a certain condition, i.e. the function we passed returns True.

```python
# example of map(), reduce() and filter().
nums = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, nums)
print("map",(list(squared)))
#-----------------------------------------
from functools import reduce
nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print("reduce",(product))
#-----------------------------------------
nums = [1, 2, 3, 4]
even = filter(lambda x: x % 2 == 0, nums)
print("filter",(list(even)))
```

Q3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

---

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def generate_lists(start, end):
    numbers = list(range(start, end + 1))
    squares = list(map(lambda x: x ** 2, numbers))
    return numbers, squares
if __name__ == "__main__":
    try:
        start_num = int(input("Enter the start number: "))
        end_num = int(input("Enter the end number: "))
        if start_num > end_num:
            print("Start number must be less than or equal to end number.")
        else:
            numbers_list, squares_list = generate_lists(start_num, end_num)
            print("Numbers:", numbers_list)
            print("Squares:", squares_list)
    except ValueError:
        print("Please enter valid integers.")
```

Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

---

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
import math
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(math.sqrt(number)) + 1):
        if number % i == 0:
            return False
    return True
num = int(input("Enter a number between 1 and 200: "))
if 1 <= num <= 200:
    if is_prime(num):
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is not a prime number.")
else:
    print("Number is out of range.")
```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

---

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current = 0
        self.next_num = 1
        self.counter = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.counter < self.num_terms:
            value = self.current
            self.current, self.next_num = self.next_num, self.current + self.next_num
            self.counter += 1
            return value
        else:
            raise StopIteration
if __name__ == "__main__":
    num_terms = 10
    fib_iter = FibonacciIterator(num_terms)
    for num in fib_iter:
        print(num, end=" ")
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

---

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2**exponent
if __name__ == "__main__":
    max_exp = 5
    power_gen = powers_of_two(max_exp)
    print(f"Powers of 2 up to exponent {max_exp}:")
    for power in power_gen:
        print(power)
```

Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
user_input = input("Enter temperatures in Celsius (comma-separated): ")
celsius = list(map(float, user_input.split(',')))
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print("Temperatures in Fahrenheit:", fahrenheit)
```
